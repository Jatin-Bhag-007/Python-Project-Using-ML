# BAR GRAPH

```python
y=df["ProductType"]
x=df["MRP"]
plt.bar(x,y,label="Mart Sales",color="red")
plt.xlabel("MRP")
plt.ylabel("ProductType")
plt.title("BIG MART SALES DATASET")
plt.show()
```

# PIE CHART

```python
slices=[50,14,24,16]
depts=["ProductID","ProductType","MRP","ProductVisibility"]
cols=["cyan","red","yellow","blue"]
plt.pie(slices,labels=depts,colors=cols,startangle=90,explode=(0,0.2,0,0),shadow=True,autopct="%.1f%%")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("big_mart.csv")
```

```python
df
```

# DISTRIBUTION PLOT

```python
sns.displot(data=df,x="MRP",bins=10,kde=True,color="green")
```

# REGRESSION PLOT

```python
sns.regplot(y="ProductVisibility",x="Weight",data=df)
```

# BOX PLOT

```python
sns.pairplot(df,hue="FatContent")
```

# SCATTER PLOT

```python
sns.scatterplot(data=df,y="OutletID",x="EstablishmentYear")
```

```python
x =  int(input("enter value of x:"))
sign = input("enter sign [+,-,*,/]:")
y = int(input("enter value of y:"))

if sign == "+":
    print(f"{x} + {y} = {x+y}")
elif sign == "-":
    print(f"{x} - {y} = {x-y}")
elif sign == "*":
    print(f"{x} x {y} = {x*y}")
elif sign == "/":
    print(f"{x} / {y} = {x/y}")
else :
    print("invalid input")
```

```python
df=df.drop_duplicates(keep="first")
```

```python
# sort the data : sort on "MPG_City" in desending order
df=df.sort_values(by="MPG_City",ascending=False)
```

```python
df
```

```python
#find out how many continents are manufacturing the car
df["Origin"].unique()
```

```python
#convert continent into numeric
#convert asia into 1, Europe into 2,USA into 3

def Origin_to_num(x):
    if x == "Asia":
        return 1
    if x == "Europe":
        return 2
    if x == "USA":
        return 3
```

```python
df['Origin']= df["Origin"].apply(Origin_to_num)
```
